```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("aggregated_metrics.csv")
sizes = df["noisy_dense_size"].unique()
df.sort_values(by="epsilon", inplace=True)
```

```python
plt.figure()
for s in sizes:
    df_s = df[df["noisy_dense_size"] == s]
    plt.plot(df_s["epsilon"], df_s["accuracy"], marker="o", label=f"size={s}")

plt.xlabel("epsilon")
plt.ylabel("accuracy")
plt.title("Accuracy vs Epsilon")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.figure()
for s in sizes:
    df_s = df[df["noisy_dense_size"] == s]
    plt.plot(df_s["epsilon"], df_s["average_mse_reconstruction"], marker="o", label=f"size={s}")

plt.xlabel("epsilon")
plt.ylabel("average_mse_reconstruction")
plt.title("Average MSE Reconstruction vs Epsilon")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.figure()
for s in sizes:
    df_s = df[df["noisy_dense_size"] == s]
    plt.plot(df_s["epsilon"], df_s["accuracy"], marker="o", label=f"accuracy (size={s})")
    plt.plot(df_s["epsilon"], df_s["average_mse_reconstruction"], marker="o", label=f"MSE (size={s})")

plt.xlabel("epsilon")
plt.ylabel("value")
plt.title("Accuracy & Average MSE vs Epsilon")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.figure()
for s in sizes:
    df_s = df[df["noisy_dense_size"] == s]
    plt.plot(df_s["epsilon"], df_s["runtime_seconds"], marker="o", label=f"size={s}")

plt.xlabel("epsilon")
plt.ylabel("runtime_seconds")
plt.title("Runtime vs Epsilon")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.figure()
for s in sizes:
    df_s = df[df["noisy_dense_size"] == s]
    plt.plot(df_s["epsilon"], df_s["loss"], marker="o", label=f"size={s}")

plt.xlabel("epsilon")
plt.ylabel("loss")
plt.title("Loss vs Epsilon")
plt.grid(True)
plt.legend()
plt.tight_layout()
```
