```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_plot = pd.read_csv("aggregated_metrics.csv")
df_plot.sort_values(by="epsilon", inplace=True)
```

```python
# Runtime vs Epsilon
plt.figure()
plt.plot(df_plot["epsilon"], df_plot["runtime_seconds"], marker="o")
plt.xlabel("epsilon")
plt.ylabel("runtime_seconds")
plt.title("Runtime vs Epsilon")
plt.grid(True)
# plt.xscale("log")  # <- optional, if you prefer a log x-axis
plt.tight_layout()

# Accuracy vs Epsilon
plt.figure()
plt.plot(df_plot["epsilon"], df_plot["accuracy"], marker="o")
plt.xlabel("epsilon")
plt.ylabel("accuracy")
plt.title("Accuracy vs Epsilon")
plt.grid(True)
# plt.xscale("log")
plt.tight_layout()

# Average MSE vs Epsilon
plt.figure()
plt.plot(df_plot["epsilon"], df_plot["average_mse_reconstruction"], marker="o")
plt.xlabel("epsilon")
plt.ylabel("average_mse_reconstruction")
plt.title("Average MSE Reconstruction vs Epsilon")
plt.grid(True)
# plt.xscale("log")
plt.tight_layout()

# Loss vs Epsilon
plt.figure()
plt.plot(df_plot["epsilon"], df_plot["loss"], marker="o")
plt.xlabel("epsilon")
plt.ylabel("loss")
plt.title("Loss vs Epsilon")
plt.grid(True)
# plt.xscale("log")
plt.tight_layout()

# Combined: Accuracy + Average MSE
plt.figure()
plt.plot(df_plot["epsilon"], df_plot["accuracy"], marker="o", label="accuracy")
plt.plot(df_plot["epsilon"], df_plot["average_mse_reconstruction"], marker="o", label="average_mse_reconstruction")
plt.xlabel("epsilon")
plt.ylabel("value")
plt.title("Accuracy & Average MSE vs Epsilon")
plt.legend()
plt.grid(True)
# plt.xscale("log")
plt.tight_layout()

plt.show()
```
